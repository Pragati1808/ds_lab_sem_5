```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency as cc
from scipy.stats import chi2

df=[[69,120,123],
    [56, 80, 52]]

st,p,dof,exp = cc(df)

print('dof=%d' %dof)
print(exp)

prob = 0.95

cr= chi2.ppf(prob,dof)

print('probab=%.3f , critical =%.3f, stat=%.3f' %(prob,cr,st))

if st>= cr:
    print('rejected H0')
else:
    print('failed to reject H0')
    
al=0.05
print('sig=%.3f, p =%.3f' %(al,p))

if p <= al:
    print('rejected H0')
else:
    print('failed to reject H0')
    
```
